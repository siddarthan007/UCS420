# Q1. Ask the user to enter a value (e.g., M) for a mathematical function.

- Generate x values from -10 to 10 using `np.linspace()`.
- Compute y values for:
$$ y = M \cdot x^2 $$  
$$ y = M \cdot \sin(x) $$
- Plot both functions in a single figure:
  - Use different colors and line styles.
  - Add legend, grid, and title.


```python
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter the value of M: "))
x = np.linspace(-10, 10)
y1 = M * x**2
y2 = M * np.sin(x)


plt.figure(figsize=(8, 6))
plt.plot(x, y1, label=r'$y = M \cdot x^2$', color='b', linestyle='-')
plt.plot(x, y2, label=r'$y = M \cdot \sin(x)$', color='r', linestyle='--')

plt.title(r'Plot of $y = M \cdot x^2$ and $y = M \cdot \sin(x)$')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

# Q2. Create a dataset of five subjects and scores.

- Convert it into a **Pandas DataFrame**.
- Plot the scores using a **Seaborn bar plot** with:
  - Different colors for each bar.
  - Annotations on top of each bar.
  - Title, axis labels, and grid.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Subjects': ['Math', 'Science', 'English', 'History', 'Computer'],
    'Scores': [73, 90, 68, 82, 95]
}

df = pd.DataFrame(data)

sns.set_style("darkgrid")

plt.figure(figsize=(8, 6))
colors = sns.color_palette("mako", len(df)) 
ax = sns.barplot(x='Subjects', y='Scores', data=df, hue='Subjects', palette=colors, legend=False)

for index, row in df.iterrows():
    ax.text(index, row.Scores + 1, str(row.Scores), ha='center', fontsize=12)

plt.title("Student Scores in Different Subjects", fontsize=14, fontweight='bold')
plt.xlabel("Subjects", fontsize=12, fontweight='bold')
plt.ylabel("Scores", fontsize=12, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

# Q3. Write a Python script to select your roll number as seed for NumPy and then generate a dataset of 50 values using `np.random.randn()`.

- Create a **2×2 subplot layout** displaying:
  - **Line plot** showing the **cumulative sum**.
  - **Scatter plot** with **random noise**.
- Customize the plots with:
  - **Titles** for each subplot.
  - **Axis labels** for better readability.
  - **Grids** for better visualization.


```python
import numpy as np
import matplotlib.pyplot as plt

roll_number = 102367010
np.random.seed(roll_number)

data = np.random.randn(50)

print("Data:", data)

fig, axes = plt.subplots(2, 1, figsize=(8, 10))

axes[0].plot(np.cumsum(data), color='b', linestyle='-', marker='o', markersize=5)
axes[0].set_title("Cumulative Sum Line Plot")
axes[0].set_xlabel("Index")
axes[0].set_ylabel("Cumulative Sum")
axes[0].grid(True)

x_values = np.arange(50)
y_values = data + np.random.randn(50) * 0.5  # Adding noise
axes[1].scatter(x_values, y_values, color='r', alpha=0.7)
axes[1].set_title("Scatter Plot with Random Noise")
axes[1].set_xlabel("Index")
axes[1].set_ylabel("Values")
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

# Q4. Data Visualization using Seaborn

### Dataset: [company_sales_data.csv](https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv)

Apply the **Seaborn** library to perform the following tasks:

1. **Line Plot**: Read the **Total Profit** of all months and display it.
2. **Multi-Line Plot**: Read **all product sales data** and show them in a single plot.
3. **Bar Chart**: Plot a **bar chart** for all features/attributes.

### Steps:
- Load the dataset using **Pandas**.
- Use **Seaborn** for data visualization.
- Customize plots with **titles, labels, grids, and legends** for clarity.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales_data.csv")

sns.set_style("darkgrid")

plt.figure(figsize=(10, 5))
sns.lineplot(x=df['month_number'], y=df['total_profit'], marker='o', color='b')
plt.title("Total Profit Over Months", fontsize=14, fontweight='bold')
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Profit", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

plt.figure(figsize=(12, 6))
for column in df.columns[1:-1]: 
    sns.lineplot(x=df['month_number'], y=df[column], marker='o', label=column)

plt.title("Sales of Different Products Over Months", fontsize=14, fontweight='bold')
plt.xlabel("Month", fontsize=12)
plt.ylabel("Units Sold", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Products")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

plt.figure(figsize=(12, 6))
df.set_index("month_number").plot(kind="bar", figsize=(12, 6), colormap="tab10", width=0.8, edgecolor="black")
plt.title("Comparison of All Features Across Months", fontsize=14, fontweight='bold')
plt.xlabel("Month", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Features", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
